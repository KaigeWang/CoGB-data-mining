```python
path = '../../Data/Clustering_results/'
```

```python
def plotPattern(name):
    global path
    df = pd.read_csv(path + name)
    totalSize = df.shape[0]
    labels = df['label'].unique()
    df['label'].value_counts()
    patterns = []
    for x in labels:
        pattern,size = utilities.generatePattern(df, x, 1)
        print('Cluster size:' + str(size))
        print('Weight:' + str(round((size/totalSize),2)*100) + '%')
        patterns.append(pattern)
    return patterns
```

```python
import pickle
```

```python
outpath = '../../Data/Temp_pattern/'
def store(Type, Season, patterns):
    with open(outpath+Type + Season + '.pickle', 'wb') as handle:
        pickle.dump(patterns, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
